# Base #1 -- Variables

**Présentement:**
* Permet de stocker <u>une</u> information à la fois<br>On verra les collections plus tard.
* Type du contenu: 
    * chaîne de caractères ou 
    * nombre (entier ou fractionnaire)

### Comment nommer les variables (et les paramètres)?

* Débuter par un **nom**, jamais un *verbe*
* Commencer par une **minuscule**
* Utiliser le format "CamelCase" si possible
* Ne pas utiliser d'accent (possible en Python, mais pas dans d'autres langages)

### Déclaration

```python
def ____________():
    |
    |
    |
    |
    |
```

* Mot-clé `def` pour définir une fonction
* Le nom de la fonction débute par
    * une minuscule
    * un verbe d'action à l'infinitif
* Parenthèses obligatoires
    * Vides si aucun paramètre à fournir
    * énumération des paramètres (noms des variables qui recevront les valeurs) séparés par une virgule
* Deux-point (`:`)
* Corps de la fonction
    * L'indentation est importante; elle permet d'identifier les lignes qui font partie de la fonction.

## Autre cas d'étude

#### Première version

```python
def formaterCitation(message, auteur):
    message = '"' + message + '",\n   ' + auteur
    return message
```

#### Deuxième  version

```python
def formaterCitation(message, auteur):
    citation = '"%s", \n   %s' % (message, auteur)
    return citation
```

* Quelle est la différence?

* Laquelle choisir? 
    * On choisira celle qui est la plus facile à comprendre
    * On évitera de modifier les **paramètres**; il faut plutôt créer une nouvelle variable


````python
def formaterCitation(message, auteur):
    message = '"' + message + '",\n   ' + auteur
    return message

print(formaterCitation('À vaincre sans péril, on triomphe sans gloire.', 'Pierre Corneille'))

def formaterCitation(message, auteur):
    citation = '"%s", \n   %s' % (message, auteur)
    return citation

print(formaterCitation('Plus vous saurez regarder loin dans le passé, plus vous verrez loin dans le futur.', 'Winston Churchill'))
````
